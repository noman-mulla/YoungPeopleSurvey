```python
def loadData(filePath):
    train_data = pd.read_csv(train_file_path)
    return train_data
```

```python
def removeRowsHavingMissingValuesinCategories(train_data):
    #train_data_reduced = train_data[train_data.[['Education', 'Lying', 'Internet usage', 'Village - town', 'Punctuality', 'Only child', 'Left - right handed', 'Smoking', 'Alcohol', 'Gender', 'House - block of flats']].dropna(axis=0, how='any')]
    train_data_reduced = train_data.dropna(subset=['Education', 'Lying', 'Internet usage', 'Village - town', 'Punctuality', 'Only child', 'Left - right handed', 'Smoking', 'Alcohol', 'Gender', 'House - block of flats'])
    
    #train_data_reduced = train_data[['Education', 'Lying', 'Internet usage', 'Village - town', 'Punctuality', 'Only child', 'Left - right handed', 'Smoking', 'Alcohol', 'Gender', 'House - block of flats']].dropna(axis=0, how='any')
    return train_data_reduced
```

```python
def getNumericColumns(train_data):
    #getting only numeric data
    train_data_numeric = train_data._get_numeric_data()
    return train_data_numeric
```

```python
def getCategoricalColumns(train_data):
    train_data_category = train_data[['Education', 'Lying', 'Internet usage', 'Village - town', 'Punctuality', 'Only child', 'Left - right handed', 'Smoking', 'Alcohol', 'Gender', 'House - block of flats']]
    return train_data_category
```

```python
def getReducedCategory(train_data):
    train_data_category_reduced = train_data[['Education', 'Lying', 'Internet usage', 'Village - town', 'Punctuality', 'Only child', 'Left - right handed', 'Smoking', 'Alcohol', 'Gender', 'House - block of flats']].dropna(axis=0, how='any')
    return train_data_category_reduced
```

```python
def getBMI(train_data_demog):
    #calculating bmi
    #getting age weight and height data
    #train_data = train_data_mix.copy()
    
    
    #train_data.drop(['Height','Weight'],axis=1)
    bmi = pd.DataFrame(train_data_demog['Weight']/((train_data_demog['Height'])**2))
    bmi['bmi_index'] = ['underweight' if x < 18.5  else 'overweight' if x > 24.9 else 'healthy' for x in bmi[0]]
    #train_data_result = train_data.join(bmi['bmi_index'])
    return bmi['bmi_index']
```

```python
# Recursive Feature Elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

def getRFEOnLinearRegression(X,Y):
    # create a base classifier used to evaluate a subset of attributes
    model = LinearRegression()
    # create the RFE model and select 3 attributes
    rfe = RFE(model, 3)
    rfe = rfe.fit(X, Y)
    # summarize the selection of the attributes
    print(rfe.support_)
    print(rfe.ranking_)
    #column names sorted by ranking
    feature_names = list(X.columns.values)
    print("Features sorted by their rank:")
    print(sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feature_names)))
```

```python
from sklearn.linear_model import LogisticRegression
def getRFEOnLogisticRegression(X,Y):
    # create a base classifier used to evaluate a subset of attributes
    model = LogisticRegression()
    # create the RFE model and select 3 attributes
    rfe = RFE(model, 3)
    rfe = rfe.fit(X, Y)
    # summarize the selection of the attributes
    print(rfe.support_)
    print(rfe.ranking_)
    #column names sorted by ranking
    feature_names = list(X.columns.values)
    print("Features sorted by their rank:")
    print(sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feature_names)))
    features_sorted = sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feature_names))
    return features_sorted
```

```python
features_sorted = getRFEOnLogisticRegression(X,Y)
```

```python
##train validation data split

from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X, Y_num,random_state = 0)
```

```python
from sklearn.decomposition import PCA
# Make an instance of the Model
pca = PCA(.95)
pca.fit(train_X)
train_X = pca.transform(train_X)
val_X = pca.transform(val_X)
```

```python
#Random Forest train
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=5, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=1,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
forest_model.fit(train_X,train_Y)
```

```python
#Random Forest predict
from sklearn.metrics import accuracy_score
forest_pred = forest_model.predict(val_X)
accuracy_score(val_Y, forest_pred)
```

```python
forest_pred_train = forest_model.predict(train_X)
accuracy_score(train_Y, forest_pred_train)
```
