```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def getScoreFromRandomForest(train_X,train_Y,val_X,val_Y):
    forest_model = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=10, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=1,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
    forest_model.fit(train_X,train_Y)
    forest_pred = forest_model.predict(val_X)
    forest_pred_train = forest_model.predict(train_X)
    print(accuracy_score(train_Y, forest_pred_train))
    print(accuracy_score(val_Y, forest_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
def getScoreFromRandomForestRegressor(train_X,train_Y,val_X,val_Y):
    regr = RandomForestRegressor(max_depth=10, random_state=0)
    regr.fit(train_X,train_Y)
    regr_pred = regr.predict(val_X)
    regr_pred_train = regr.predict(train_X)
    #print("{0}--{1}".format(regr_pred_train, train_Y))
    print(r2_score(train_Y, regr_pred_train))
    print(r2_score(val_Y, regr_pred))
```

```python
from sklearn.svm import LinearSVC
def getScoreLinearSVC(train_X,train_Y,val_X,val_Y):
    clf = LinearSVC(C=10,random_state=0,multi_class="ovr")
    clf.fit(train_X,train_Y)
    clf_pred = clf.predict(val_X)
    clf_pred_train = clf.predict(train_X)
    print(accuracy_score(train_Y, clf_pred_train))
    print(accuracy_score(val_Y, clf_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
def getScoreNaiveBayes(train_X,train_Y,val_X,val_Y):
    clf = MultinomialNB()
    clf.fit(train_X,train_Y)
    clf_pred = clf.predict(val_X)
    clf_pred_train = clf.predict(train_X)
    print(accuracy_score(train_Y, clf_pred_train))
    print(accuracy_score(val_Y, clf_pred))
```

```python
from sklearn.model_selection import train_test_split

def splitDataAndTestOnModel(X,Y,model):
    train_X, val_X, train_Y, val_Y = train_test_split(X, Y,random_state = 0,test_size=0.33)
    if(model == 'SVC'):
        getScoreLinearSVC(train_X,train_Y,val_X,val_Y)
    if(model == 'Naive'):
        getScoreLinearSVC(train_X,train_Y,val_X,val_Y)
    if(model == 'Random Regressor'):
        getScoreFromRandomForestRegressor(train_X,train_Y,val_X,val_Y)
    if(model == 'Random'):
        getScoreFromRandomForest(train_X,train_Y,val_X,val_Y)
```
